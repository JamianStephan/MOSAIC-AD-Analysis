```python
import dispersion_analysis_moffat as moff_AD
import dispersion_analysis_durham as dur_AD
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
analysis=dur_AD.AD_simulation()
analysis.load_ZA([0,15,30])
analysis.load_MOSAIC_band("LR_VIS_G",.01)
analysis.calculate_snapshot_shifts(.51)
analysis.load_PSFs()
analysis.load_aperture()
analysis.calculate_snapshot_transmissions()
analysis.snapshot_plots()
```

```python
analysis=dur_AD.AD_simulation()
analysis.load_HA(1.5,2.5,-60.5)
analysis.load_MOSAIC_band("LR_VIS_G",.01)
analysis.calculate_integration_shifts(.69,.51)
analysis.load_PSFs()
analysis.load_aperture()
analysis.calculate_integration_transmissions()
analysis2=dur_AD.AD_simulation()
analysis2.load_HA(1.5,2.5,-60.5)
analysis2.load_MOSAIC_band("LR_NIR_All",.01)
analysis2.calculate_integration_shifts(.69,1.04)
analysis2.load_PSFs()
analysis2.load_aperture()
analysis2.calculate_integration_transmissions()

plt.figure(figsize=[7,5])
plt.ylim(0,1.1)
plt.plot(analysis.output['wavelengths'],analysis.output['relative_integration_transmissions'],color='blue',linewidth=1)
plt.ylim(0,1.1)
plt.xlabel("Wavelength (nm)")
plt.title("HDF-S ({} deg), HA: {}-{}h, NIR All + VIS {}".format(analysis.input['declination'],analysis.input['HA_range'][0],analysis.input['HA_range'][-1],"G"))
plt.plot(analysis2.output['wavelengths'],analysis2.output['relative_integration_transmissions'],color='red',linewidth=1)


analysis=moff_AD.AD_simulation()
analysis.load_HA(1.5,2.5,-60.5)
analysis.load_MOSAIC_band("LR_VIS_G",.01)
analysis.calculate_integration_shifts(.71,.51)
analysis.load_PSFs()
analysis.load_aperture()
analysis.calculate_integration_transmissions()
analysis2=moff_AD.AD_simulation()
analysis2.load_HA(1.5,2.5,2.-60.5)
analysis2.load_MOSAIC_band("LR_NIR_All",.01)
analysis2.calculate_integration_shifts(.71,1.04)
analysis2.load_PSFs()
analysis2.load_aperture()
analysis2.calculate_integration_transmissions()
        
plt.plot(analysis.output['wavelengths'],analysis.output['relative_integration_transmissions'],color='blue',linewidth=1,linestyle='-.')
plt.plot(analysis2.output['wavelengths'],analysis2.output['relative_integration_transmissions'],color='red',linewidth=1,linestyle='-.')
plt.axhline(y=0,label='Durham',linewidth=0.9)
plt.axhline(y=0,label='Moffat',linewidth=0.9,linestyle='-.')
plt.axhline(y=1,label='No AD Transmission',color='black',linestyle='--')
plt.axvline(x=analysis.input['guide_waveref'],color='black',linewidth=0.7,label='Guide WL = {}um'.format(analysis.input['guide_waveref'],2))
plt.legend()
```

```python
analysis=dur_AD.AD_simulation(HA_samples=5)
analysis.load_HA(1.5,2.5,2.2)
analysis.load_MOSAIC_band("LR_VIS_G",.01)
analysis.calculate_integration_shifts(.71,.51)
analysis.load_PSFs()
analysis.load_aperture()
analysis.calculate_integration_transmissions()
analysis.integration_plots()
plt.title("VIS-G Integration Shifts, HA = {}-{}h, Dec = {} deg".format(analysis.input['HA_range'][0],analysis.input['HA_range'][-1],analysis.input['declination']))
plt.savefig("int track VIS.png",dpi=400,bbox_inches='tight')
analysis2=dur_AD.AD_simulation(HA_samples=5)
analysis2.load_HA(1.5,2.5,2.2)
analysis2.load_MOSAIC_band("LR_NIR_All",.01)
analysis2.calculate_integration_shifts(.71,1.04)
analysis2.load_PSFs()
analysis2.load_aperture()
analysis2.calculate_integration_transmissions()
analysis2.integration_plots()
plt.title("NIR-All Integration Shifts, HA = {}-{}h, Dec = {} deg".format(analysis.input['HA_range'][0],analysis.input['HA_range'][-1],analysis.input['declination']))
plt.figure(figsize=[7,5])
plt.axhline(y=1,label='No AD Transmission, {}'.format("numerical durham"),color='black',linestyle='--')
plt.axvline(x=analysis.input['guide_waveref'],color='black',linewidth=0.7,label='Guide WL = {}um'.format(analysis.input['guide_waveref'],2))
plt.plot(analysis.output['wavelengths'],analysis.output['relative_integration_transmissions'],label="Aperture Centre = {}um at 2h".format(analysis.input['aperture_waveref']))
plt.ylim(0,1.1)
plt.xlabel("Wavelength (nm)")
plt.ylabel("Transmission Relative to No-AD")
plt.title("VIS-G + NIR-All Integration Transmission, Dec: {} deg (COSMOS), HA: {}-{}h".format(analysis.input['declination'],analysis.input['HA_range'][0],analysis.input['HA_range'][-1],"G"))

plt.plot(analysis2.output['wavelengths'],analysis2.output['relative_integration_transmissions'],label="Aperture Centre = {}um at 2h".format(analysis2.input['aperture_waveref']))
plt.legend()
plt.axvline(x=analysis.input['aperture_waveref'],color='C0',linewidth=0.7)
plt.axvline(x=analysis2.input['aperture_waveref'],color='C1',linewidth=0.7)
```
