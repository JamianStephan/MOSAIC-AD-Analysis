```python
import sys
sys.path.append('../')

import dispersion_analysis_durham as AD_dur
import dispersion_analysis_moffat as AD_mof
import numpy as np
import matplotlib as mpl
import dispersion_V3
import pandas as pd
import scipy
from scipy import integrate

def integ_metric(transmission):

    metric=[]
    metric.append(min(transmission))
    metric.append(max(transmission))
    metric.append(scipy.integrate.simpson(transmission)/(len(transmission)-1))

    return metric

VIS_B=[.390,.458]
VIS_V=[.450,.591] 
VIS_R=[.586,.770]

VIS_I=[.390,.554]
VIS_II=[.549,.780]
VIS_All=[.390,.780]
NIR_All=[.770,1.857]
```

```python
decs = np.arange(-80,25,5)
HAs = np.arange(0,5,1)
VIS_bands=["B","G","R"]
PSFs=['moffat']

total_decs=[]
for p in range(0,len(PSFs)):
    for o in range(0,len(VIS_bands)):
        for dec in decs:
            for i in range(0,len(HAs)):         
                total_decs.append(dec)
total_HAs=[]
for p in range(0,len(PSFs)):
    for o in range(0,len(VIS_bands)):   
        for i in range(0,len(decs)):
            for HA in HAs:
                total_HAs.append(HA)
total_bands=[]
for p in range(0,len(PSFs)):
    for band in VIS_bands:
        for i in range(0,len(decs)):
            for o in range(0,len(HAs)):
              total_bands.append(band)
total_PSFs=[]
for PSF in PSFs:
    for o in range(0,len(VIS_bands)):
        for i in range(0,len(HAs)):
            for p in range(0,len(decs)):
                total_PSFs.append(PSF)  

data = {'HA mid': total_HAs,'Dec': total_decs,'VIS band': total_bands,'PSF': total_PSFs}
metrics=pd.DataFrame(data)

metrics['Guide']=np.nan
metrics['VIS Centre']=np.nan
metrics['NIR Centre']=np.nan

metrics['VIS Min']=np.nan
metrics['VIS Max']=np.nan
metrics['VIS Throughput']=np.nan

metrics['NIR Min']=np.nan
metrics['NIR Max']=np.nan
metrics['NIR Throughput']=np.nan

metrics['Metric']=np.nan

metrics.to_csv("V3_moffat_refined.csv",index=False)
```

```python
indexes=np.arange(92,315,1)
for i in indexes:
        metrics=pd.read_csv("generic_test_cases_refined.csv")
        print(i)
        targ_dec=metrics.at[i,'Dec']

        centring="mid HA"
        VIS_band=metrics.at[i,'VIS band']
        VIS_centre=metrics.at[i,'VIS Centre']
        NIR_centre=metrics.at[i,'NIR Centre']

        guide=metrics.at[i,'Guide']

        analysis=AD_dur.AD_simulation()
        analysis.load_HA(metrics.at[i,'HA mid']-0.5,metrics.at[i,'HA mid']+0.5,targ_dec)
        analysis.load_MOSAIC_band("LR_VIS_"+str(VIS_band),.002)
        analysis.calculate_integration_shifts(guide,VIS_centre)
        analysis.load_PSFs()
        analysis.load_aperture()
        analysis.calculate_integration_transmissions()
        VIS_metrics=integ_metric(analysis.output['relative_integration_transmissions'])
        
        analysis=AD_dur.AD_simulation()
        analysis.load_HA(metrics.at[i,'HA mid']-0.5,metrics.at[i,'HA mid']+0.5,targ_dec)
        analysis.load_MOSAIC_band("LR_NIR_All",.01)
        analysis.calculate_integration_shifts(guide,NIR_centre)
        analysis.load_PSFs()
        analysis.load_aperture()
        analysis.calculate_integration_transmissions()
        NIR_metrics=integ_metric(analysis.output['relative_integration_transmissions'])    

        metrics=pd.read_csv("V3_durham_refined.csv")
        metrics.at[i,'Guide']=guide
        metrics.at[i,'VIS Centre']=VIS_centre
        metrics.at[i,'NIR Centre']=NIR_centre
        metrics.at[i,'VIS Min']=VIS_metrics[0]
        metrics.at[i,'VIS Max']=VIS_metrics[1]
        metrics.at[i,'VIS Throughput']=VIS_metrics[2]
        metrics.at[i,'NIR Min']=NIR_metrics[0]
        metrics.at[i,'NIR Max']=NIR_metrics[1]
        metrics.at[i,'NIR Throughput']=NIR_metrics[2]
        metrics.to_csv("V3_durham_refined.csv",index=False)
    
```
